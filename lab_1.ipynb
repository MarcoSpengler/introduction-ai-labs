```python
from pandas.plotting import scatter_matrix

attributes = ["median_house_value", "median_income", "total_rooms",
              "housing_median_age", "longitude", "latitude",	"total_bedrooms",	"population",	"households"]
scatter_matrix(housing[attributes], figsize=(12, 8))
```

```python
#check if sklearn is installed
import sklearn
from sklearn.model_selection import train_test_split

# Split the dataset into a training and test set, with 20% of the data in the test set
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

# Check the length of the training and test set
print(f"Training Set: {len(train_set)} entries")
print(f"Test Set: {len(test_set)} entries")
```

```python
# create 5 income categories based on the median income of the districts

import numpy as np

housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])

housing["income_cat"].hist()
```
